```python
from ipywidgets import interact
from bokeh.plotting import figure as bf
from bokeh.layouts import layout as bl
from bokeh.models import Toggle, BoxAnnotation, CustomJS, ColumnDataSource, LabelSet
from bokeh.io import push_notebook, show, output_notebook
output_notebook()
```

# Original Minimal example

```python
p = bf(title='test', x_range=(0,1), y_range=(0,1))
x = [1/3, 2/3]
y = [1/3, 2/3]
labels = ['label_%i' % i for i in range(len(x))]
p.circle(x=x, y=y, size=15)

box = BoxAnnotation(left=None, right=0.5, fill_color='red', fill_alpha=0.1)

p.add_layout(box)

labels = ['label_%i' % i for i in range(len(x))]
source = ColumnDataSource(data={'x': x, 'y': y, 'labels': labels})
labelset = LabelSet(x='x', y='y', text='labels', source=source, text_alpha = 0, background_fill_alpha=1)
p.add_layout(labelset)

code = '''\
if toggle.active
    box.visible = true
    labelset.visible = true
    console.log 'enabling box'
else
    box.visible = false
    labelset.visible = false
    console.log 'disabling box'
'''
callback = CustomJS.from_coffeescript(code=code, args={})
toggle = Toggle(label="Red Box", button_type="success", callback=callback)
callback.args = {'toggle': toggle, 'box': box, 'labelset': labelset}

layout = bl([p], [toggle])
show(layout)
```

# Working code

```python
p = bf(title='test', x_range=(0,1), y_range=(0,1))
x = [1/3, 2/3]
y = [1/3, 2/3]
labels = ['label_%i' % i for i in range(len(x))]
p.circle(x=x, y=y, size=15)

box = BoxAnnotation(left=None, right=0.5, fill_color='red', fill_alpha=0.1)

p.add_layout(box)

labels = ['label_%i' % i for i in range(len(x))]
source = ColumnDataSource(data={'x': x, 'y': y, 'labels': labels})
labelset = LabelSet(x='x', y='y', text='labels', source=source)
p.add_layout(labelset)

code = '''\
if toggle.active
    box.fill_alpha = 0.1
    box.line_alpha = 1
    labelset.text_alpha = 1
    console.log 'enabling box 2'
else
    box.fill_alpha = 0
    box.line_alpha = 0
    
    labelset.text_alpha = 0
    console.log 'disabling box 2'
'''
callback = CustomJS.from_coffeescript(code=code, args={})
toggle = Toggle(label="Red Box", button_type="success", callback=callback)
callback.args = {'toggle': toggle, 'box': box, 'labelset': labelset}

layout = bl([p], [toggle])
show(layout)
```
